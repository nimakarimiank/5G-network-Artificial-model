```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
"""
Implementation of GFA network with Blind Procedure
List of Variables:

"""
def data_generator_gfa(number_of_stations,number_of_slots,Lambda):
    station_slot_package = np.zeros([number_of_stations,number_of_slots],dtype=int)
    for i in range(0,number_of_stations):
        poisson_distributed_over_numberofslots=np.clip(np.random.poisson(Lambda,number_of_slots),a_min=0,a_max=1)
        station_slot_package[i,:]=poisson_distributed_over_numberofslots
    station_slot_package.astype(int)
    return station_slot_package


def export_dataSet(number_of_stations,number_of_slots,Lambda):
    generated_sample = data_generator_gfa(number_of_stations,number_of_slots,Lambda)
    tensor_generarated = tf.convert_to_tensor(generated_sample,dtype=tf.int32)

    data_frame = pd.DataFrame(generated_sample)
    tensor_dataframe = pd.DataFrame(tensor_generarated)
    tensor_dataframe.to_excel('Tensor_FinalDataset.xlsx',index=True,index_label='Station ID')
    data_frame.to_excel('FinalDataset.xlsx',index=True,index_label='Station ID')
```

```python
export_dataSet(10,20,0.1)
```
